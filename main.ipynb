# 1. Introduction

This project aims in detecting stress levels on the basis of a scale (1-10) of teenage students due to academics, sports, etc. 

This project aims to help identify and reduce stress levels in our peers, aidinf in their quality of life, happiness, clarity, studies, etc.

Dataset used can be found in this repository under the name `survey_responses.csv`, data was collected by a survey conducted via google forms, with 63 responses as of writing this. You may find the survey here: `https://tinyurl.com/mrynwnvk`.
